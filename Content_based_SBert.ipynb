```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from pyvi import ViTokenizer, ViPosTagger
```

```python
# viết hàm tiền xử lý dữ liệu trên cột overview

def Preprocessing(data):
  # stop_words = set(stopwords.words('english'))
  f = open("/content/drive/MyDrive/Ky1nam4/stopwords/vietnamese.txt", "r")
  stop_words = set(f.read())
#  - chuyển về từ thường
  ii = 1
  for i in range(len(data)):
#  - Xóa dấu câu, ký tự đặc biệt
#  - xóa stopword
#  - tách từ
#  ...
    if ii % 1000 == 0:
      print(ii)
    ii += 1
    data[i] = re.sub('\W+', ' ', data[i].lower())
    tokens = ViTokenizer.tokenize(data[i])
    word_tokens = tokens.strip().split(' ')
    data[i] = [w for w in word_tokens if not w in stop_words]
    data[i] = ' '.join(data[i][:100])
```

```python
Preprocessing(books.description)
```
