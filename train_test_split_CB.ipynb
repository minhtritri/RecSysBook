## Làm tại đây

```python
books['all_time_quantity_sold'] = books['all_time_quantity_sold'].fillna(0)
```

```python
books.info()
```

```python
df_id_list = ratings.groupby('user_id')['product_id'].apply(list).reset_index(name='list_product')
df_id_list
```

```python
list_temp = df_id_list[df_id_list.user_id == 27.0]['list_product'][0]
len(list_temp)
```

```python
len(set(list_temp))
```

```python
for i in range(len(books.id)):
  rating_mean = ratings[ratings.product_id == books.id[i]].groupby('product_id')['rating_average'].mean().reset_index().rating_average[0]
  books['rating_average'][i] = rating_mean
  print("so thu {}, gia tri {}".format(i,rating_mean))
```

```python
books.rating_average.value_counts()
```

```python
books_temp = books[books.id.isin(list_temp)]
books_temp = books_temp.sort_values(by = ['rating_average', 'review_count', 'all_time_quantity_sold'], ascending = False)
books_temp.head(10)
```

```python
df_id_list
```

```python
df_id_list[df_id_list.user_id == 27.0]['list_product'][0]
```

```python
from sklearn.model_selection import train_test_split
test_list = []
train_list = []
user_id_list = []
for id in df_id_list.user_id:
  for i in df_id_list[df_id_list.user_id == id]['list_product']:
    # print(i)
    books_temp = books[books.id.isin(i)]
    books_temp = books_temp.sort_values(by = ['rating_average', 'review_count', 'all_time_quantity_sold'], ascending = False)
    X_train, X_test = train_test_split(books_temp.id, test_size=0.3, random_state=42)
    train_list.append(list(X_train))
    test_list.append(list(X_test))
    user_id_list.append(id)
    print("Id: {}, co length train: {}, test: {}".format(id, len(X_train), len(X_test)))
```

```python
test_list
```

```python

  
# dictionary of lists 
dict = {'user_id': user_id_list, 'train_list': train_list, 'test_list': test_list} 
    
df = pd.DataFrame(dict)
```
