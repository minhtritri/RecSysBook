```python
train_test = pd.read_csv('/content/drive/MyDrive/Ky1nam4/DoAnCuoiKy/Data_final/data_final/content-based/train_test_CB')
train_test
```

```python
vectors = pd.read_csv('/content/drive/MyDrive/Ky1nam4/DoAnCuoiKy/Data_final/CB-sbert.csv', header = None)
```

```python
vectors = vectors.to_numpy()
```

```python
def Average(lst):
    return sum(lst) / len(lst)
```

### MRR

```python
result_list = []
for index in range(len(train_test.user_id)):
  vector_total = np.zeros(768)
  # print(id)
  list_train = eval(train_test[train_test.user_id == train_test.user_id[index]]
                    .train_list[index])
  len_train = len(list(books[books.id.isin(list_train)].index))
  for i in list(books[books.id.isin(list_train)].index):
    vector_total += vectors[i]
  vector_train = vector_total/len_train
  list_test = eval(train_test[train_test.user_id == train_test.user_id[index]]
                   .test_list[index])
  vector_test = vectors[list(books[books.id.isin(list_test)].index)]

  similarity_matrix = cosine_similarity(vector_train.reshape(1, -1), vector_test)
  result = 0.0
  for i in range(len(similarity_matrix[0])):
    if similarity_matrix[0][i] >= 0.5:
      result = 1/(i+1)
      break
      if i == len(similarity_matrix[0])-1:
        result = 0
    
  result_list.append(result)
print("Ket qua MRR: {}".format(round(Average(result_list), 2)))
```

### NDCG

```python
vector_total = np.zeros(768)
# print(id)
list_train = eval(train_test[train_test.user_id == 27.0].train_list[0])
len_train = len(list(books[books.id.isin(list_train)].index))
for i in list(books[books.id.isin(list_train)].index):
  vector_total += vectors[i]
vector_train = vector_total/len_train
list_test = eval(train_test[train_test.user_id == 27.0]
                  .test_list[0])
vector_test = vectors[list(books[books.id.isin(list_test)].index)]

similarity_matrix = cosine_similarity(vector_train.reshape(1, -1), vector_test)
```

```python
similarity_matrix[0].tolist()
```

```python
dict = {'relevance': similarity_matrix[0].tolist()} 
    
df = pd.DataFrame(dict)
```

```python
df['log2'] = np.log2(df.index + 2)
df['chia'] = df['relevance']/df['log2']
df['DCG'] = df['chia']

for i in range(len(df.DCG)):
  if i == 0:
    continue
  else:
    df['DCG'][i] += df['DCG'][i-1]
df['relevance'] = df['relevance'].sort_values(ascending=False).reset_index(drop = True)
df['chia2'] = df['relevance']/df['log2']
df['IDCG'] = df['chia2']
for i in range(len(df.IDCG)):
  if i == 0:
    continue
  else:
    df['IDCG'][i] += df['IDCG'][i-1]

df['NDCG'] = df['DCG']/df['IDCG']
df['NDCG'].mean()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from pyvi import ViTokenizer, ViPosTagger
```

```python
# viết hàm tiền xử lý dữ liệu trên cột overview

def Preprocessing(data):
  # stop_words = set(stopwords.words('english'))
  f = open("/content/drive/MyDrive/Ky1nam4/stopwords/vietnamese.txt", "r")
  stop_words = set(f.read())
#  - chuyển về từ thường
  ii = 1
  for i in range(len(data)):
#  - Xóa dấu câu, ký tự đặc biệt
#  - xóa stopword
#  - tách từ
#  ...
    if ii % 1000 == 0:
      print(ii)
    ii += 1
    data[i] = re.sub('\W+', ' ', data[i].lower())
    tokens = ViTokenizer.tokenize(data[i])
    word_tokens = tokens.strip().split(' ')
    data[i] = [w for w in word_tokens if not w in stop_words]
    data[i] = ' '.join(data[i][:100])
```
