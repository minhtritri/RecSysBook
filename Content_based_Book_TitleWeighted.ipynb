```python
#đánh index cho các bộ phim bằng  pd.Series() và lưu trong biến mapping
mapping = pd.Series(books.index, index = books.name)
mapping
```

```python
#viết hàm nhận vào index của bộ phim trả về tên của sách
def get_book_title(bookId):
  return books.name.iloc[bookId]
```

```python
#viết hàm trả và top 10 sách sẽ đem đi khuyến nghị cho người dùng U. Người dung U đã xem bộ phim có tên là ""
def get_recommendation(movie_input):
  movie_index = mapping[movie_input]

  similarity_movie = list(enumerate(similarity_matrix[movie_index]))
  similarity_movie.sort(key = lambda item: item[1], reverse = True)

  relate_movie = [get_book_title(bookid) for (bookid, similarity) in similarity_movie[:11] if bookid != 1]
  return relate_movie
```
