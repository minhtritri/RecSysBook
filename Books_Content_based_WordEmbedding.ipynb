```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from pyvi import ViTokenizer, ViPosTagger
```

```python
# viết hàm tiền xử lý dữ liệu trên cột overview

def Preprocessing(data):
  # stop_words = set(stopwords.words('english'))
  f = open("/content/drive/MyDrive/Ky1nam4/stopwords/vietnamese.txt", "r")
  stop_words = set(f.read())
#  - chuyển về từ thường
  ii = 1
  for i in range(len(data)):
#  - Xóa dấu câu, ký tự đặc biệt
#  - xóa stopword
#  - tách từ
#  ...
    if ii % 1000 == 0:
      print(ii)
    ii += 1
    data[i] = re.sub('\W+', ' ', data[i].lower())
    tokens = ViTokenizer.tokenize(data[i])
    word_tokens = tokens.strip().split(' ')
    data[i] = [w for w in word_tokens if not w in stop_words]
    data[i] = ' '.join(data[i][:100])
```

```python
Preprocessing(books.description)
```

```python
from keras.preprocessing.text import Tokenizer
tok = Tokenizer(filters='')
tok.fit_on_texts(books.description)


vocab_size = len(tok.word_index)
print(tok.word_index)
print(vocab_size)


encd_cont = tok.texts_to_sequences(books.description)
print(encd_cont)
```

```python
max_cont_len=100  # Chiều dài lớn nhất trong train
vocab_size = len(tok.word_index) + 1  # tổng số từ trong từ vựng
embed_dim = 300 # chiều dài embedding
```

```python
# now padding to have a amximum length of 19
from keras_preprocessing.sequence import pad_sequences
pad_cont= pad_sequences(encd_cont, maxlen=max_cont_len, padding='post')
pad_cont.shape
```

```python
def get_key(val):
  for key, value in tok.word_index.items():
      if val == value:
        return key
      
  return 0
get_key(108)
```

```python
from gensim.models import KeyedVectors
from gensim import models

word2vec_path = '/content/drive/MyDrive/Ky1nam4/DoAnCuoiKy/Data_final/word_embedding/baomoi.window2.vn.model.bin.gz'
w2v_model = models.KeyedVectors.load_word2vec_format(word2vec_path, binary=True)
```

```python
#Combine 
total_d = []
for i in range(len(pad_cont)):
  print(i)
  # set_lamda = 5
  total_d_tu = np.zeros(300)
  total_d_mau = 200
  for word in range(len(pad_cont[i])):
    # print(get_key(pad_cont[1][word]))
    try: 
      V = w2v_model.wv.get_vector(get_key(pad_cont[i][word])) #check trong w2v nếu có
    except:
      V = np.zeros(300)#check trong w2v nếu ko có thì V = vector 0
    # if get_key(pad_cont[i][word]) in books.title[i]:
    #   lamda = set_lamda
    # else:
    #   lamda = 1
    # try:
    #   tfidf_val = df_tfidf[get_key(pad_cont[i][word])][i]
    # except:
    #   tfidf_val = 0
    total_d_tu += V  # tử số
  total_d.append(total_d_tu/(total_d_mau))
      
```
